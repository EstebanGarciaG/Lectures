# Data Curation

**Data curation is a series of processes that ensuredata accessibility, quality, and usability for research and analysis.**


![data_curation.png](merge_df.png)


## Critical Roles of Data Curation:

### Data Quality Assurance
- Ensures the accuracy, completeness, and reliability of data.
- Facilitates the identification and correction of errors.

### Enhanced Data Accessibility
- Makes data easily accessible to researchers and policymakers.
- Supports timely decision-making during health crises.

### Data Integration
- Allows for the combination of data from various sources.
- Enhances the comprehensiveness of the analysis.
- Having harmonization rules.

### Improved Data Security
- Protects sensitive information from unauthorized access.
- Ensures compliance with ethical and legal standards.

### Facilitation of Data Sharing
- Encourages collaboration among researchers and institutions.
- Promotes transparency and reproducibility in research.

### Support for Advanced Analytics
- Provides clean and well-structured data for sophisticated analytical methods.
- Enables the use of machine learning and AI in predictive modeling.

### Resource Optimization
- Reduces the time and effort required to manage data.
- Allows researchers to focus on analysis and interpretation.

```python
import requests
import pandas as pd
```

```python
df['age'].loc[df['ageu']==1]=df['age']/365
df['ageu'].loc[df['ageu']==1]=3
plt.figure(figsize=(10, 6))
plt.boxplot(df['age'].dropna())
plt.title('Age Distribution')
plt.xlabel('Patients')
plt.ylabel('Age')
plt.show()
```

```python
df = df.loc[(df['age'] > 0) & (df['age'] < 120)]
plt.figure(figsize=(10, 6))
plt.boxplot(df['age'].dropna())
plt.title('Age Distribution')
plt.xlabel('Patients')
plt.ylabel('Age')
plt.show()
```

```python
print('Number of rows: ',len(df))
print('Number of Patients: ',len(df['subjid'].unique()))
```

```python
duplicated_ids = df[df.duplicated('subjid', keep=False)].copy()
duplicated_ids
```

```python
# Set non-matching 'sex' to NaN in duplicated_ids
duplicated_ids.loc[duplicated_ids.duplicated('subjid', keep=False) & (duplicated_ids.duplicated(['subjid', 'sex'], keep=False) == False), 'sex'] = float('nan')

# Set non-matching 'age' to NaN in duplicated_ids
duplicated_ids.loc[duplicated_ids.duplicated('subjid', keep=False) & (duplicated_ids.duplicated(['subjid', 'age'], keep=False) == False), 'age'] = float('nan')

# Set non-matching 'pregnancy' to NaN in duplicated_ids
duplicated_ids.loc[duplicated_ids.duplicated('subjid', keep=False) & (duplicated_ids.duplicated(['subjid', 'pregnancy'], keep=False) == False), 'pregnancy'] = float('nan')
duplicated_ids=duplicated_ids.drop_duplicates()
duplicated_ids
```

```python
cleaned_duplicated_df = duplicated_ids.dropna(subset=['age', 'sex'], how='all')
cleaned_duplicated_df
```

```python
df_1 = df[~df['subjid'].isin(duplicated_ids['subjid'])]
df_1 = pd.concat([df_1, cleaned_duplicated_df], ignore_index=True)
print('Number of rows: ',len(df_1))
print('Number of Patients: ',len(df_1['subjid'].unique()))
df_1.sample(5)
```

## Data Integration
- **Schema Matching:** Aligning different data sources to a common schema.
- **Appending Data:** Combining datasets by stacking one on top of another, resulting in a longer dataset with more rows.
- **Joining Data:** Connecting related data points by adding datasets side by side, resulting in a wider dataset with more columns.


```python
df_2 = pd.read_csv('/content/patient_level_data2.csv')
df_2.sample(5)
```

```python
df2_map=df_2.copy()
sex_mapping = {'M': 1, 'F': 2}
df2_map['sex_at_birth'] = df2_map['sex_at_birth'].map(sex_mapping)

smoking_mapping = {'Current': 1, 'Former': 1,'Never':0}
df2_map['smoking'] = df2_map['smoking'].map(smoking_mapping)

column_mapping = { 'sex_at_birth': 'sex', 'age(years)': 'age'}
df2_map.rename(columns=column_mapping, inplace=True)

df2_map.insert(df2_map.columns.get_loc('age') + 1, 'ageu', 3)

df2_map.sample(5)
```


## Data Transformation
- **Aggregation:** Summarizing data to provide a consolidated view.
- **Filtering:** Selecting relevant subsets of data based on criteria.
- **Encoding:** Converting categorical data into numerical format.


```python
api_df.sample(8)
```

```python
api_df['Final_Date'] = api_df['CompletionDate'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
api_df['Country'] = api_df['LocationCountry'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
api_df['Intervention'] = api_df['InterventionName'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)

api_df[['Country','Final_Date','Intervention']]

api_df['Year_Month'] = pd.to_datetime(api_df['Final_Date'], errors='coerce').dt.to_period('M').astype(str)

# Handle the cases where the date format is 'Month Year'
mask = api_df['Year_Month'] == 'NaT'
api_df.loc[mask, 'Year_Month'] = pd.to_datetime(api_df.loc[mask, 'Final_Date'], format='%B %Y', errors='coerce').dt.to_period('M').astype(str)

api_df_interest=api_df[['Country','Year_Month','Intervention']]
api_df_interest_aggr=api_df_interest.groupby(['Country','Year_Month']).count().reset_index()
api_df_interest_aggr=api_df_interest_aggr.loc[api_df_interest_aggr['Intervention']>0]
api_df_interest_aggr
```

```python
url = "http://api.worldbank.org/v2/country?format=json&per_page=300"

# Fetch the data
response = requests.get(url)
data = response.json()

# Extract relevant information
countries_info = pd.json_normalize(data[1])
income_classification = countries_info[['name', 'incomeLevel.value']]
income_classification.sample(5)
```

merge_df.png

```python
api_df_interest_aggr_income=pd.merge(api_df_interest_aggr,income_classification,left_on='Country',right_on='name',how='inner')
api_df_interest_aggr_income['Year'] = api_df_interest_aggr_income['Year_Month'].str.split('-').str[0]
api_df_interest_aggr_income['Month'] = api_df_interest_aggr_income['Year_Month'].str.split('-').str[1]
api_df_interest_aggr_income.sample(5)
```

```python
api_df_interest_aggr_income=api_df_interest_aggr_income.loc[api_df_interest_aggr_income['Year']=='2022']
grouped_df = api_df_interest_aggr_income.groupby(['incomeLevel.value', 'Year_Month']).size().unstack(fill_value=0)

# Plotting the number of interventions by income level and month
plt.figure(figsize=(25, 6))
grouped_df.T.plot(kind='bar', stacked=True)
plt.title('Number of Interventions by Income Level and Month')
plt.xlabel('Year_Month')
plt.ylabel('Number of Interventions')
plt.legend(title='Income Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Good Practices
- **Documentation:** For transparency, reproducibility, and standardization.
- **Accuracy Check:** Verifying that data correctly represents the real-world entities or phenomena.
- **Reusable Functions:** Package code into functions that are reusable for different datasets.
- **Clear Communication:** Be clear about limitations, caveats, and assumptions.
- **Standardize Values:** Use standardized codes (e.g., SNOMED codes for M and Female) for sharing and analysis, ISO codes for countries.
- **Consistency Check:** Ensuring uniformity of data across datasets.
- **Missing Data Handling:** Filling in or excluding missing data points.
